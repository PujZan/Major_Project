```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
# Load the CSV file
csv_path = "C:\\Users\\Acer\\Desktop\\Model\\Dataset.csv"  # Replace with your CSV file path
data = pd.read_csv(csv_path)

# Assuming the CSV has a column 'Label'
label_column = 'Label'  # Replace with the actual column name for labels

# Calculate total number of datasets (samples) and total number of classes
total_datasets = len(data)
total_classes = data[label_column].nunique()

# Print total number of datasets and classes
print(f"Total number of datasets (samples): {total_datasets}")
print(f"Total number of classes: {total_classes}")

# Calculate class distribution
class_distribution = data[label_column].value_counts()

# Plot class distribution where count > 1000
class_distribution_gt_1000 = class_distribution[class_distribution > 1000]
plt.figure(figsize=(10, 6))
ax = class_distribution_gt_1000.plot(kind='bar', color='skyblue')
plt.title("Class Distribution (Count > 1000)")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate each bar with the number of images
for i, count in enumerate(class_distribution_gt_1000):
    ax.text(i, count + 50, str(count), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Plot class distribution where 400 < count <= 1000
class_distribution_500_to_1000 = class_distribution[(class_distribution > 400) & (class_distribution <= 1000)]
plt.figure(figsize=(10, 6))
ax = class_distribution_500_to_1000.plot(kind='bar', color='lightgreen')
plt.title("Class Distribution (400 < Count <= 1000)")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate each bar with the number of images
for i, count in enumerate(class_distribution_500_to_1000):
    ax.text(i, count + 20, str(count), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Filter classes with counts greater than 2000
class_distribution_gt_2000 = class_distribution[class_distribution > 2000]

# Print the list of classes
class_list = class_distribution_gt_2000.index.tolist()
print("Classes with counts greater than 2000:")
print(class_list)

# Total dataset count
total_samples = 112120

# Function to calculate percentage of total samples
def absolute_value(val):
    total = sum(class_distribution_gt_2000)
    return f'{val / 100 * total:.0f} ({val:.1f}%)'

# Plot class distribution as a pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    class_distribution_gt_2000, 
    labels=class_distribution_gt_2000.index, 
    autopct=lambda p: absolute_value(p),
    startangle=140, 
    colors=plt.cm.tab20.colors
)
plt.title("Class Distribution (Count > 2000) out of Total Dataset (112,120)")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.tight_layout()
plt.show()
```
